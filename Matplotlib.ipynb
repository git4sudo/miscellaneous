4.  **Given Below is the Air quality index(AQI) of Delhi and Mumbai.Find how many are good(0-50),moderate(51-100),unhealthy(101-150)**
AQI_Delhi=[45,66,89,47,89,150,134,145,150,125,130,150]

      AQI_Mumbai=[40,34,67,47,45,134,100,120,150,100,90,80]

```python
AQI_Delhi = [45,66,89,47,89,150,134,145,150,125,130,150]
AQI_Mumbai = [40,34,67,47,45,134,100,120,150,100,90,80]
AQI_Quality_Count = [[0,0],[0,0],[0,0]]                                #[[good(0-50)][moderate(51-100)][unhealthy(101-150)]]                            #good(0-50),moderate(51-100),unhealthy(101-150)
for i in AQI_Delhi:
    if i >= 0 and i<= 50:
        AQI_Quality_Count[0][0]+=1
    elif i >= 51 and i<= 100:
        AQI_Quality_Count[1][0]+=1
    else:
        AQI_Quality_Count[2][0]+=1
for i in AQI_Mumbai:
    if i >= 0 and i<= 50:
        AQI_Quality_Count[0][1]+=1
    elif i >= 51 and i<= 100:
        AQI_Quality_Count[1][1]+=1
    else:
        AQI_Quality_Count[2][1]+=1
        
        
labels = ['AQI_Delhi', 'AQI_Mumbai']
width = 0.50
good_AQI = np.array(AQI_Quality_Count[0])
moderate_AQI = np.array(AQI_Quality_Count[1])
unhealthy_AQI = np.array(AQI_Quality_Count[2])

fig, ax = plt.subplots()
ax = fig.add_axes([0,0,1,1])
ax.bar(labels, good_AQI, width, label='Good', color = 'g')
ax.bar(labels, moderate_AQI, width, bottom=good_AQI,label='Moderate', color = 'y')
ax.bar(labels, unhealthy_AQI, width, bottom=good_AQI+moderate_AQI,label='Unhealthy', color = 'r')

ax.set_ylabel('Air quality index(AQI) Count')
ax.set_xlabel('Cities')
ax.set_title('Compare AQI of Delhi and Mumbai')
ax.legend()
plt.show()
```

5. **Find the value of y for each value of x using the equation Y=5x+10 where x=[2,4,7,1,5,9,12,10].Represent the value of X and Y using the scatter plot**

```python
#write the code here

x=[2,4,7,1,5,9,12,10]
X = np.array(x)
Y = 5*X + 10
print('Calculated Y from X: ',X,'->',Y)
plt.scatter(X, Y, alpha=0.75, color = '#99E6B2',marker='o', edgecolors='#2C1171',s=100)
plt.xlabel('Given X values')
plt.ylabel('Y = 5x+10 ')
plt.title('X vs Y scatter plot')
plt.show()
```

**6.  Draw the three different sub plots with the following equation. Plot x values on the x-axis and y-values on the y-axis. Give the neccessary labels in the plot. You can select x values of your choice. calculate y for atleast 50 values of x**

y1 = x^2 + 3x + 4 <br>

y3 = logx <br>

```python
x = np.linspace(1.0, 100.0, num=50)

y1 = x**2 + 3*x + 4 
y2 = 1/(1 + np.exp(-x))
y3 = np.log(x)


plt.subplot(3, 1, 1)
plt.plot(x, y1, '*-')
plt.title('Sub plots for Y1 - Y2 - Y3')
plt.ylabel('y1 = x^2 + 3x + 4')
plt.subplot(3, 1, 2)
plt.plot(x, y2, '.-')
plt.ylabel('y2 = sigmoid(x)')
plt.subplot(3, 1, 3)
plt.plot(x, y3, 'o-')
plt.xlabel('Uniformly spaced number of step 50')
plt.ylabel('y3 = logx')
plt.show()
```

```python
data = pd.read_csv("/content/scatter_plot_data.csv")
ratio = data['ratio']
Y = np.log(np.array(data['view_count']).tolist())
Y_likes = np.log(np.array(data['likes']).tolist())
Y_bin = np.arange(11.5, 19, 0.5).tolist()
Y_likes_bin = np.arange(4, 16, 0.5).tolist()
Ratio_Count = np.arange(0, 100.5, 0.5).tolist()
Y_Dislike = np.log(np.array((100.0-np.array(data['ratio']))/ratio) * np.array(data['likes'])).tolist()
```

```python
plt.hist(Y,bins = Y_bin, color = 'green', edgecolor = 'black', alpha=0.5)
plt.xlabel('Distribution of View Counts - log(View_Count)')
plt.ylabel('Number of View Counts')
plt.title(r'Histogram of View Count')
plt.show()

plt.hist(Y_likes,bins = Y_likes_bin, color = 'blue', edgecolor = 'black', alpha=0.5)
plt.hist(Y_Dislike,bins = Y_likes_bin, color = 'red', edgecolor = 'black', alpha=0.5)
plt.legend(['Likes','Dislikes'])
plt.xlabel('Distribution of Like & Dislike Counts')
plt.ylabel('Number of Like Counts')
plt.title(r'Histogram of Like & Dislike Count')
plt.show()

plt.hist(ratio,range = [0,100],bins=Ratio_Count, color = 'purple', edgecolor = 'black', alpha=0.5)
plt.xlabel('Ratio of Likes from 0 to 100%')
plt.ylabel('Number of Ratio of Likes')
plt.title(r'Histogram of Ratio of Likes')
plt.show()
```

2.   In figure 2 the distribution of 'likes' is shifted to left than the distribution of 'dislikes' which means on an average people are putting more like than dislike on a youtube video. 
3.   In figure 3, according to the plot very few youtube videos have like to dislike ratio less than 60%, which means for almost all the videos the number of like is greater than the dislike.
