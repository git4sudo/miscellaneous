```python
AQI_Before = [17,40,13,13,15,37,7,14,14,19,27,61]
AQI_After = [5,9,8,8,0,12,4,6,9,18,28,60]

labels = ['PM2.5','PM10','NO','NO2','NOx','NH3','CO','SO2','O3','Benzene','Toluene','Xylene']
width = 0.5
fig, ax = plt.subplots()

fig.set_figwidth(10)
fig.set_figheight(10)

ax = fig.add_axes([0,0,1,1])
ax.bar(labels, AQI_Before, width, label='Period of Observation Before Lockdown', color = '#FF6961')
ax.bar(labels, AQI_After, width, bottom=AQI_Before,label='Period of Observation After Lockdown', color = '#A7C7E7')
ax.tick_params(axis='x', labelrotation=0)
plt.xticks(ha='center')
ax.set_ylabel('Air quality index(AQI) Missing Count')
ax.set_xlabel('Pollutants')
ax.set_title('Ptg of Missing Values')
ax.legend()
plt.draw()
```

```python
ordered_df = dft1.sort_values(by='Absolute_Decrease_Mean_Concentration')
my_range=range(1,len(dft1.index)+1)

f = plt.figure()
f.set_figwidth(25)
f.set_figheight(10)

plt.hlines(y=my_range, xmin=0, xmax=ordered_df['Absolute_Decrease_Mean_Concentration'], color='#A7C7E7')
plt.plot(ordered_df['Absolute_Decrease_Mean_Concentration'], my_range, "o",color='#A7C7E7')

for index, value in enumerate(ordered_df['Absolute_Decrease_Mean_Concentration']):
    plt.text(value, index+1,'   ['+str(value)+']',horizontalalignment='left',verticalalignment='center')

plt.yticks(my_range, ordered_df['Pollutant'])
plt.title("Absolute Decrease Mean Concentration Value of Pollutants", loc='center')
plt.xlabel('Absolute Decrease Mean Concentration Value')
plt.ylabel('Pollutants')

plt.show()
```

```python
ordered_df = dft1.sort_values(by='Percentage_Decrease_Mean_Concentration')
my_range=range(1,len(dft1.index)+1)

f = plt.figure()
f.set_figwidth(25)
f.set_figheight(10)

plt.hlines(y=my_range, xmin=0, xmax=ordered_df['Percentage_Decrease_Mean_Concentration'], color='#FF6961')
plt.plot(ordered_df['Percentage_Decrease_Mean_Concentration'], my_range, "o",color='#FF6961')

for index, value in enumerate(ordered_df['Percentage_Decrease_Mean_Concentration']):
    plt.text(value, index+1,'   ['+str(value)+']',horizontalalignment='left',verticalalignment='center')

plt.yticks(my_range, ordered_df['Pollutant'])
plt.title("Percentage Decrease Mean Concentration Value of Pollutants", loc='center')
plt.xlabel('Percentage Decrease Mean Concentration Value')
plt.ylabel('Pollutants')

plt.show()
```
